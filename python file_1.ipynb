```python
import numpy as np
import pandas as pd
from yaml import safe_load
import os
from tqdm import tqdm
import pickle
```

```python
filename = []

for file in os.listdir('t20s'):
    filename.append(os.path.join('t20s',file))
```

```python
filename[0:5]
```

```python
final_data = pd.DataFrame()
counter=0
for file in tqdm(filename):
    with open(file,'r') as f:
        df = pd.json_normalize(safe_load(f))
        df['match_id'] = counter
        final_data=final_data.append(df)
        counter+=1
```
