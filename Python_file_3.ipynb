```python
'''
Required Data
1. Batting team - Done
2. Balling team - Done
3. city         - Done
4. current score - Done
5. score in last 5 over
6. wicket left 
7. CRR
8. Balls left
9. current score
'''
```

```python
data['Current_score'] = data.groupby('match_id').cumsum()['runs']
```

```python
data
```

```python
#balls bowled or balls left 
data['over'] = data['ball_of_match'].apply(lambda x:str(x).split(".")[0])
data['ball'] = data['ball_of_match'].apply(lambda x:str(x).split('.')[1])
data
```

```python
data['Total_balls'] = (data['over'].astype('int')*6) + data['ball'].astype('int')
data
```

```python
data['Balls_left'] = 120-data['Total_balls']
data['Balls_left'] = data['Balls_left'].apply(lambda x:0 if x<0 else x)
data
```

```python
'''
Required Data
1. Batting team - Done
2. Balling team - Done
3. city         - Done
4. current score - Done
5. Balls left    - Done
7. score in last 5 over 
8. wicket left 
9. CRR 
'''
```

```python

data['player_dismissed'] = data['player_dismissed'].apply(lambda x: 0 if x=='0' else 1)
data
```

```python
data['player_dismissed'] = data.groupby('match_id').cumsum()['player_dismissed']
data['Wicket_remaining'] = 10-data['player_dismissed']
data
```

```python
'''
Required Data
1. Batting team - Done
2. Balling team - Done
3. city         - Done
4. current score - Done
5. Balls left    - Done
8. wicket left  -  Done
9. CRR           -  Done
7. score in last 5 over -  Done
'''
```

```python
data['CRR'] = (data['Current_score']*6)/data['Total_balls']
data
```

```python
#last five overs score
groups = data.groupby('match_id')

match_ids = data['match_id'].unique()
last_five = []
for id in match_ids:
    last_five.extend(groups.get_group(id).rolling(window=30).sum()['runs'].values.tolist())
```

```python
data['Last_5_over_score'] = last_five
```

```python
data = data.groupby('match_id').sum()['runs'].reset_index().merge(data,on='match_id')
```

```python
data = data[['match_id','batting_team','bowling_team','Current_score','Balls_left','Wicket_remaining',
                   'CRR','Last_5_over_score','city','runs_x']]
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
data = data.sample(data.shape[0])
```

```python
data
```

```python
x = data.drop(columns=['runs_x' , 'match_id'])
y = data['runs_x']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder 
from sklearn.pipeline import Pipeline 
from sklearn.preprocessing import StandardScaler 
from sklearn.ensemble import RandomForestRegressor 
from xgboost import XGBRegressor 
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
pipe.fit(x_train,y_train)
y_pred = pipe.predict(x_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
pickle.dump(pipe,open('train.pkl','wb'))
```
